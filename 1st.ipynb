```python
smalldata = data[['artist.inverted', 'track', 'time', 'date.entered',
        'x1st.week', 'x2nd.week', 'x3rd.week',]]
smalldata
```

```python
smelt= smalldata.melt(['artist', 'track', 'time', 'date.entered'], ['wk1', 'wk2', 'wk3'], 'weeks','rate' )
smelt
```

```python
smelt.head()
```

```python
smelt['w'] = smelt['weeks'].str.slice(2)
```

```python
smelt['w'] = smelt['weeks'].apply (lambda x : int(x[2]))
```

```python
smelt
```

```python
smelt['date.entered'][0]+pd.to_timedelta('7days')
smelt
```
