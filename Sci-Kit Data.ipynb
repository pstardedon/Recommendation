### Graphical Representation

```python
plt.figure(figsize = [20,8])
plt.hist(dff1['PRICE'], bins =50, ec='r', width = 0.1)
plt.grid()
plt.show()
```

```python
plt.figure(figsize = [20,8])
plt.hist(dff1['CRIM'], bins =50, ec='r', width = 0.8)
plt.grid()
plt.show()
```

```python
import seaborn as sns
```

```python
# plt.figure(figsize = [3,3])
# sns.pairplot(dff1['CRIM'], bins =50, ec='r', width = 0.8)
sns.pairplot(data= dff1)
plt.grid()
plt.show()
```

```python
# plt.figure(figsize = [3,3])
# sns.pairplot(dff1['CRIM'], bins =50, ec='r', width = 0.8)
sns.distplot(dff1['PRICE'])
plt.grid()
plt.show()
```

```python
# plt.figure(figsize=(16,10))
sns.jointplot(x = "RM", y = "PRICE", data = dff1, kind = 'reg', size = 8)
plt.grid()
plt.show()
```
