```python
z = []
for files in glob.glob('*.csv'):  
    z.append(files)
    
```

```python
newData = pd.DataFrame()
for items in z:
    df = pd.read_csv(items)
    newData = pd.concat([newData, df])
```

```python
newData.dropna(inplace = True)
```

```python
bestData = newData[newData['Quantity Ordered'] != 'Quantity Ordered']
```

```python
bestData['Quantity Ordered'] = pd.to_numeric(bestData['Quantity Ordered'])
bestData['Price Each'] = pd.to_numeric(bestData['Price Each'])
```

```python
bestData['Amount'] = bestData['Price Each']*bestData['Quantity Ordered']
```

```python
bestData
```

```python
bestData['Month'] = bestData['Order Date'].str[0:2]
```

```python
bestData['Month'] = pd.to_numeric(bestData['Month'])
```

```python
grpData = bestData.groupby('Month')['Amount']
```

```python
y = grpData.sum()
```

```python
x = range(1,13)

plt.bar(x,grpData.sum())
plt.xticks(x)
plt.title('Sales Analysis')
plt.xlabel('Month')
plt.ylabel('Sales')
plt.show()
```

```python
x = range(1,13)

plt.plot(x,y, 'r-o')
plt.xticks(x)
plt.title('Sales Analysis')
plt.xlabel('Month')
plt.ylabel('Sales')
plt.grid()
plt.show()
```

```python
x = range(1,13)
plt.scatter(x,y)
plt.plot(x,y, 'r-o')
plt.xticks(x)
plt.title('Sales Analysis')
plt.xlabel('Month')
plt.ylabel('Sales')
plt.grid()
plt.show()

plt.show()
```

```python
def getCity(address):
    return address.split(',')[1]

def getState(address):
    return address.split(',')[2].split(' ')[1]

bestData['City'] = bestData['Purchase Address'].apply(lambda x: getCity(x))
bestData['State'] = bestData['Purchase Address'].apply(lambda x: getState(x))
```
