```python
dat = pd.read_csv('all_student.csv')
```

```python
data = dat
```

```python
data.columns = (['Data'])

data
```

```python
data['Serial'] = data['Data'].apply(lambda x: x.split(' | ')[0])
data['First Name'] = data['Data'].apply(lambda x: x.split(' | ')[1])
data['Last Name'] = data['Data'].apply(lambda x: x.split(' | ')[2])
data['Age'] = data['Data'].apply(lambda x: x.split(' | ')[3])
data['State'] = data['Data'].apply(lambda x: x.split(' | ')[4])
data['Department'] = data['Data'].apply(lambda x: x.split(' | ')[5])
data['Level'] = data['Data'].apply(lambda x: x.split(' | ')[6])
data['Gender'] = data['Data'].apply(lambda x: x.split(' | ')[7])
```

```python
cleanData = data
```

```python
cleanData.drop('Data', inplace = True, axis = 1)
```

```python
cleanData['Age'] = pd.to_numeric(cleanData['Age'])
```

```python
cleanData[cleanData['Age']<21]
```

```python
cleanData['Serial'] = cleanData['Serial'].str[2:4]
cleanData['State'] = cleanData['State'].astype('string')
```
